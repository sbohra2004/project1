```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/archive (15).zip')
df
```

```python


# Check for missing values
print(df.isnull().sum())    #missing values and number of missing values


# Check for duplicate rows
print(df.duplicated().sum())

# Data types of columns
print(df.dtypes)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram for glucose
sns.histplot(df['Glucose'])
plt.title('Glucose Distribution')
plt.show()

# Boxplot for BMI
sns.boxplot(x=df['BMI'])
plt.title('BMI Distribution')
plt.show()

# Boxplot for Age
sns.boxplot(x=df['Age'])
plt.title('Age Distribution')
plt.show()
```

```python
# Count plot for Outcome (target variable)
sns.countplot(x='Outcome', data=df)    #frequency
plt.title('Outcome Distribution (Diabetes)')
plt.show()
```

```python
# Boxplot to compare Glucose levels by Outcome
sns.boxplot(x='Outcome', y='Glucose', data=df)
plt.title('Glucose Levels by Outcome comparison')
plt.show()

# Boxplot for BMI by Outcome
sns.boxplot(x='Outcome', y='BMI', data=df)
plt.title('BMI Distribution by Outcome')
plt.show()
```

```python
# Correlation matrix
correlation = df.corr()


# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True,cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

```
